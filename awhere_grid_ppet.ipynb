```python
import os

import geopandas as gpd
import matplotlib.pyplot as plt

import earthpy as et
```

```python
# Create path to grid geojson
grid_path = os.path.join(et.io.HOME, "Desktop", "drought", "data", "created", "aoi_polys.geojson")

# Read as gdf
aoi_grid = gpd.read_file(grid_path)

# Add centroid column 
aoi_grid['centroid'] = aoi_grid.geometry.apply(lambda poly: poly.centroid)
```

```python
# Create test grid with 3 points
aoi_test_grid = aoi_grid[:3]
aoi_test_grid
```

```python
# Add ppet to aoi_test_grid
aoi_test_grid = aoi_test_grid.assign(ppet=0)
aoi_test_grid
```
